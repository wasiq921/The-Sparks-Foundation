```python
#Let's make a profit vs loss graph. In order to do that we need to make a new column in the dataset.
PL = pd.Series([],dtype=pd.StringDtype())
for i in range(len(df)):
    if df['Profit'][i] > 0:
        PL[i] = 'Profit'
    else:
        PL[i] = 'Loss'

        
df.insert(loc = 11, column = 'PL', value = PL)
```

```python
sns.countplot('PL', data=df)
```

In total we have more Profit ratio than Loss

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Category', y = 'Sales', data = df )
```

Now we can see that we have more sales in Tech Category rather than Furniture and Office Supplies are vey low in Sales

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Region', y = 'Profit', hue = 'Category' , data = df )
```

We can see that we have less profit ratio in every region in Furniture and Office Supplies Category

```python
sns.histplot(x=df['Ship Mode'], color='g')
plt.title('Ship Mode Preference')
```

```python
sns.countplot(x=df['Segment'])
plt.title('Customer Segments')
```

```python
ds =df.groupby('Category')['Profit','Sales'].agg('sum')
print(ds)
ds.plot.bar()
plt.legend(loc='upper left')
plt.title('Category-wise Profit and Sale')
```

```python
plt.figure(figsize=(10,10))
df['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.title('Quantity of different Sub-Categories Ordered')
```

```python
plt.figure(figsize=(13,8))
sns.set(style="whitegrid", color_codes=True)
pal = sns.color_palette("colorblind", len(df))
sns.barplot(x='PL', y = 'Sales', data = df, palette = pal, hue='Sub-Category')
```
